# Import Pacotes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Gerando amostras

```python
parametros_normal_loc = 0
parametros_normal_scale = 1.0
size = 100

dados_normal = np.random.normal(loc=parametros_normal_loc, scale=parametros_normal_scale, size=size)

print(dados_normal)
```

```python
parametros_exponencial_scale = 2.0

dados_exponencial = np.random.exponential(scale=parametros_exponencial_scale, size=size)

print(dados_exponencial)
```

```python
parametros_uniform_low = -1
parametros_uniform_high = 1

dados_uniform = np.random.uniform(low=parametros_uniform_low, high=parametros_uniform_high, size=size)

print(dados_uniform)
```

# Estatísticas básicas

```python
dados_normal_media = dados_normal.mean(); #print(f"Média dados normal: {dados_normal_media}")
dados_normal_mediana = np.median(dados_normal); #print(f"Mediana dados normal: {dados_normal_mediana}")
dados_normal_var = dados_normal.var(); #print(f"Variância dados normal: {dados_normal_var}")
dados_normal_std = dados_normal.std(); #print(f"Desvio dados normal: {dados_normal_std}")

estatisticas_normal = {"Média": round(dados_normal_media,2),
                        "Mediana": round(dados_normal_mediana,2),
                        "Variancia": round(dados_normal_var,2),
                        "Desvio": round(dados_normal_std,2)}

print(f"\n Estatísticas Normal: {estatisticas_normal}")
```

```python
dados_exponencial_media = dados_exponencial.mean(); #print(f"Média dados Exponencial: {dados_exponencial_media}")
dados_exponencial_mediana = np.median(dados_exponencial); #print(f"Mediana dados normal: {dados_exponencial_mediana}")
dados_exponencial_var = dados_exponencial.var(); #print(f"Variância dados Exponencial: {dados_exponencial_var}")
dados_exponencial_std = dados_exponencial.std(); #print(f"Desvio dados Exponencial: {dados_exponencial_std}")

estatisticas_exponencial = {"Média": round(dados_exponencial_media,2),
                        "Mediana": round(dados_exponencial_mediana,2),
                        "Variancia": round(dados_exponencial_var,2),
                        "Desvio": round(dados_exponencial_std,2)}

print(f"\n Estatísticas Exponencial: {estatisticas_exponencial}")
```

```python
dados_uniform_media = dados_uniform.mean(); #print(f"Média dados Uniform: {dados_uniform_media}")
dados_uniform_mediana = np.median(dados_uniform); #print(f"Mediana dados normal: {dados_uniform_mediana}")
dados_uniform_var = dados_uniform.var(); #print(f"Variância dados Uniform: {dados_uniform_var}")
dados_uniform_std = dados_uniform.std(); #print(f"Desvio dados Uniform: {dados_uniform_std}")

estatisticas_uniform = {"Média": round(dados_uniform_media,2),
                        "Mediana": round(dados_uniform_mediana,2),
                        "Variancia": round(dados_uniform_var,2),
                        "Desvio": round(dados_uniform_std,2)}

print(f"\n Estatísticas Uniform: {estatisticas_uniform}")
```

```python
dicionario = {"Normal": 
                  {"params": {"loc": parametros_normal_loc,
                              "scale": parametros_normal_scale},
                   "dados": dados_normal,
                   "estatisticas": estatisticas_normal},
              
             "Exponencial": 
                  {"params": {"scale": parametros_exponencial_scale},
                   "dados": dados_exponencial,
                   "estatisticas": estatisticas_exponencial},
              
             "Uniforme": 
                  {"params": {"low": parametros_uniform_low,
                              "high": parametros_uniform_high}, 
                   "dados": dados_uniform,
                   "estatisticas": estatisticas_uniform}}

# print(dicionario)
```

```python
from pprint import pprint

for nome, dados in dicionario.items():
    print(f"{nome}:")
    pprint(dados, sort_dicts=False)
    print()
```

# Novas distribuições

## Poisson

```python
parametros_possion_lambda = 5

dados_possion = np.random.poisson(lam = parametros_possion_lambda, size=size)

print(dados_possion)
```

```python
dados_possion_media = dados_possion.mean(); #print(f"Média dados normal: {dados_normal_media}")
dados_possion_mediana = np.median(dados_possion); #print(f"Mediana dados normal: {dados_normal_mediana}")
dados_possion_var = dados_possion.var(); #print(f"Variância dados normal: {dados_normal_var}")
dados_possion_std = dados_possion.std(); #print(f"Desvio dados normal: {dados_normal_std}")

estatisticas_poisson = {"Média": round(dados_possion_media,2),
                        "Mediana": round(dados_possion_mediana,2),
                        "Variancia": round(dados_possion_var,2),
                        "Desvio": round(dados_possion_std,2)}

print(f"\n Estatísticas Poisson: {estatisticas_poisson}")
```

## Binomial

```python
parametros_binomial_p = 0.5
parametros_binomial_n = 5

dados_binomial = np.random.binomial(n = parametros_binomial_n , p = parametros_binomial_p, size=size)

print(dados_binomial)
```

```python
dados_binomial_media = dados_binomial.mean(); #print(f"Média dados normal: {dados_normal_media}")
dados_binomial_mediana = np.median(dados_binomial); #print(f"Mediana dados normal: {dados_normal_mediana}")
dados_binomial_var = dados_binomial.var(); #print(f"Variância dados normal: {dados_normal_var}")
dados_binomial_std = dados_binomial.std(); #print(f"Desvio dados normal: {dados_normal_std}")

estatisticas_binomial = {"Média": round(dados_binomial_media,2),
                        "Mediana": round(dados_binomial_mediana,2),
                        "Variancia": round(dados_binomial_var,2),
                        "Desvio": round(dados_binomial_std,2)}

print(f"\n Estatísticas Binomial: {estatisticas_binomial}")
```

# Df com distribuicoes

```python
data_completo = {
    'Normal': (estatisticas_normal),
    'Exponencial': (estatisticas_exponencial),
    'Uniforme':  (estatisticas_uniform),
    'Poisson': (estatisticas_poisson),
    'Binomial': (estatisticas_binomial)
}

df = pd.DataFrame(data_completo)
df
```

# Estatisticas adicionais

```python
distribuicoes = {
    'Normal': dados_normal,
    'Exponencial': dados_exponencial,
    'Uniforme': dados_uniform,
    'Poisson': dados_possion,
    'Binomial': dados_binomial
}
#print(distribuicoes)

stats_adicionais = {}

for nome, dados in distribuicoes.items():
    stats_adicionais[nome] = {
        'Min': round(np.min(dados),2),
        'Q1 (25%)': round(np.percentile(dados, 25),2),
        'Q2 (50%)': round(np.percentile(dados, 50),2),
        'Q3 (75%)': round(np.percentile(dados, 75),2),
        'Max': round(np.max(dados),2),
    }

stats_adicionais
```

```python
df_adicional = pd.DataFrame(stats_adicionais)
df_adicional
```

# Comparando distribuicoes

```python
dados_boxplot = [
    dados_normal, 
    dados_exponencial, 
    dados_uniform,
    dados_possion,
    dados_binomial
]

labels = ['Normal', 'Exponencial', 'Uniforme', 'Poisson', 'Binomial']

fig, ax = plt.subplots(figsize=(10, 6))

bp = ax.boxplot(
    dados_boxplot,
    labels=labels,
    patch_artist=True,
    medianprops=dict(color='black', linewidth=2)
)

ax.set_title('Comparação das Distribuições - Boxplot', fontsize=14, fontweight='bold')
ax.set_xlabel('Distribuição', fontsize=12)
ax.set_ylabel('Valores', fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Estatisticas VS. Valores Teóricos

```python
medias_teoricas = {
    'Normal':      parametros_normal_loc,
    'Exponencial': parametros_exponencial_scale,
    'Uniforme':    (parametros_uniform_low + parametros_uniform_high) / 2,
    'Poisson':     parametros_possion_lambda,
    'Binomial':    parametros_binomial_n * parametros_binomial_p
}

medias_amostra = {
    'Normal': round(dados_normal.mean(),2),
    'Exponencial': round(dados_exponencial.mean(),2),
    'Uniforme': round(dados_uniform.mean(),2),
    'Poisson': round(dados_possion.mean(),2),
    'Binomial': round(dados_binomial.mean(),2)
}

df_comparacao = pd.DataFrame({
    'Media Teorica':  medias_teoricas,
    'Media Amostral': medias_amostra
})

df_comparacao['Diferenca'] = (
    df_comparacao['Media Amostral'] - df_comparacao['Media Teorica']
    ).round(4)

df_comparacao
```
