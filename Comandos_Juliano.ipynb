# Comandos Úteis
---

## Terminal

| Comando | O que faz |
|---|---|
| `cd nome_pasta` | Entra na pasta indicada |
| `cd ..` | Volta para a pasta anterior |
| `cd ~` | Vai direto para a pasta home |
| `ls` | Lista arquivos visíveis da pasta atual |
| `ls -a` | Lista todos os arquivos, incluindo os ocultos (começam com `.`) |
| `pwd` | Mostra o caminho completo da pasta atual |
| `mkdir nome` | Cria uma nova pasta |
| `touch arquivo.txt` | Cria um arquivo vazio |
| `ctrl + shift + v` | Cola links/texto no terminal |
| `ctrl + c` | Interrompe o processo em execução |

## Git

### Configuração inicial
| Comando | O que faz |
|---|---|
| `git config --global user.name "Nome"` | Define o nome do usuário no commit |
| `git config --global user.email "email"` | Define o email do usuário no commit |

### Fluxo principal
| Comando | O que faz |
|---|---|
| `git clone <link>` | Clona o repositório da nuvem para a máquina |
| `git status` | Mostra arquivos modificados, adicionados ou removidos |
| `git diff` | Mostra as diferenças entre o repositório local e a main |
| `git add .` | Adiciona todas as alterações para o próximo commit |
| `git add arquivo.py` | Adiciona apenas um arquivo específico |
| `git commit -m "mensagem"` | Registra as alterações com uma descrição |
| `git push` | Envia os commits para a nuvem (atualiza a main) |
| `git pull` | Baixa a versão mais recente da nuvem para a máquina |

> **Ordem do fluxo:** `add` → `commit -m` → `push`

## UV

| Comando | O que faz |
|---|---|
| `wget -qO- https://astral.sh/uv/install.sh | sh` | Instala o Python
| `uv init` | Inicializa o ambiente UV na pasta atual |
| `uv add pandas matplotlib jupyterlab notebook` | Instala as bibliotecas no ambiente |
| `uv run jupyter notebook` | Abre o Jupyter Notebook no navegador |
| `uv run python script.py` | Executa um script Python no ambiente UV |

### .gitignore
| Comando | O que faz |
|---|---|
| `touch .gitignore` | Cria o arquivo `.gitignore` |

O `.gitignore` lista arquivos e extensões que o Git deve **ignorar** ao usar `git add .`.
Exemplos comuns de entradas:
```
.venv/
__pycache__/
*.csv
*.ipynb_checkpoints
```
